# Example 1 — Load Shorkie_LM and run masked-token inference\nShorkie_LM is a masked DNA language model (`use_bert=true`). Given a 16,384 bp window with one\nposition masked, it predicts the probability of A/C/G/T at that position.\n\n**Prereqs:** `pip install -e .`; `data/download.sh --models lm` (or run on the cluster). Model: `models.shorkie_lm`.

```python
# Resolve the released model directory via shorkie.config (works on the training
# cluster). External users: run `data/download.sh --models all` and set
# MODEL_OVERRIDE to that local dir, or edit config/paths.yaml:work_root.
import os, json, numpy as np
from shorkie import config
MODEL_OVERRIDE = os.environ.get("SHORKIE_MODELS")  # e.g. ./data_local/models
import pysam
from baskerville import seqnn
from shorkie.models.ensemble import make_input   # builds the (16384,170) input (DNA + species ch.114)

lm_dir = MODEL_OVERRIDE or str(config.path("models.shorkie_lm"))
params_file = f"{lm_dir}/train/params.json"
model_file  = f"{lm_dir}/train/model_best.h5"
fasta = pysam.Fastafile(str(config.path("genome.fasta")))
print("LM dir:", lm_dir)
```

### Load the language model\nThe LM input has 170 channels: 4 DNA one-hot + 166 species-identity (S. cerevisiae = channel 114).

```python
params = json.load(open(params_file))
params["model"]["num_features"] = 170          # 4 DNA + 166 species
import io, contextlib
model = seqnn.SeqNN(params["model"])
with contextlib.redirect_stdout(io.StringIO()):   # silence per-weight restore logs
    model.restore(model_file, trunk=False, by_name=False)
keras_model = model.model
print("output shape:", keras_model.output_shape)   # (None, 16384, 4) — per-position base probs
```

### Masked prediction at a chosen position\nWe mask the center of a window and ask the model to reconstruct the base.

```python
chrom, start = "chrI", 70000
end = start + 16384
center = 16384 // 2                       # masked position (0-based within the window)
x = make_input(fasta, chrom, start, end, seq_len=16384, mask_pos=center)   # DNA zeroed at `center`

probs = keras_model(x.numpy()[None, ...], training=False).numpy()[0, center, :4]
bases = ["A", "C", "G", "T"]
genome_base = fasta.fetch(chrom, start + center, start + center + 1).upper()
print("predicted P(A,C,G,T):", dict(zip(bases, np.round(probs, 4))))
print("argmax prediction:", bases[int(np.argmax(probs))], " | genome base:", genome_base)
```

`probs` is a softmax over the 4 bases. The argmax is the model's reconstruction of the masked base; the per-base log-likelihood-ratio `log(P_alt/P_ref)` is the LM's variant-effect score (used for the eQTL benchmark, `scripts/04_analysis/shorkie/eqtl/2_variant_scoring/score_variants_LM.py`).
